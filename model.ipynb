```python
# import libraries
import pandas as pd
import numpy as np



```

```python
# read and load dataset as dataframe
data_frame = pd.read_csv('laptopPrice.csv')
data_frame.head(11)
```

```python
#print unique values
for column in data_frame.columns:
    print(data_frame[column].value_counts())
    print()
```

```python
#remove non useful data
# rating,Number of Ratings,Number of Reviews

data_frame = data_frame.drop(columns=['rating', 'Number of Ratings', 'Number of Reviews'])
data_frame.sample(10)
```

```python
# for processor_gnrtn column replace Not Available with NaN
data_frame['processor_gnrtn'] = data_frame['processor_gnrtn'].replace('Not Available', np.nan)
data_frame.sample(10)
```

```python
# converting os_bit to 0 and 1
model_dataframe = data_frame
model_dataframe['os_bit'] = data_frame['os_bit'].replace('32-bit', 0)
model_dataframe['os_bit'] = data_frame['os_bit'].replace('64-bit', 1)


# converting Touchscreen to 0 and 1
model_dataframe['Touchscreen'] = data_frame['Touchscreen'].replace('No', 0)
model_dataframe['Touchscreen'] = data_frame['Touchscreen'].replace('Yes', 1)


# coverting msoffice to 0 and 1
model_dataframe['msoffice'] = data_frame['msoffice'].replace('No', 0)
model_dataframe['msoffice'] = data_frame['msoffice'].replace('Yes', 1)

print(model_dataframe.sample(10))



# co
```

```python
encoded_df = pd.get_dummies(model_dataframe, columns=['brand', 'processor_brand', 'processor_name', 'processor_gnrtn', 'ram_gb', 'ram_type', 'ssd' , 'hdd' , 'os' , 'graphic_card_gb' , 'weight' , 'warranty'  ])
encoded_df = encoded_df.astype(int)
model_df = encoded_df
pd.set_option('display.max_columns', None)
print(model_df.sample(10))
```

```python
X = model_df.drop(columns=['Price'])
Y = model_df['Price']
```

```python
from sklearn.model_selection import train_test_split


X_train , X_test , Y_train , Y_test = train_test_split(X , Y , test_size = 0.1 , random_state = 2)
```

```python
#linear Regression 
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train , Y_train)

model.score(X_test , Y_test)
```

```python
#logistic Regression 
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train , Y_train)

model.score(X_test , Y_test)
```

```python
#decision tree 
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor()
model.fit(X_train , Y_train)

model.score(X_test , Y_test)
```

```python
# random forest

from sklearn.ensemble import RandomForestRegressor

criterations = ["squared_error", "absolute_error", "friedman_mse", "poisson"]


max_fetures = ["sqrt", "log2", None]


for citeration in criterations:
    for max_feature in max_fetures:

        model = RandomForestRegressor(criterion=citeration, max_features=max_feature)
        model.fit(X_train, Y_train)

        print(citeration, max_feature)

        print(model.score(X_train, Y_train))
        print(model.score(X_test, Y_test))
        print()
```

```python
model = RandomForestRegressor(criterion="absolute_error", max_features=0.75 , n_estimators=150,random_state=3,max_samples=0.4 , max_depth=25)
model.fit(X_train, Y_train)

print(model.score(X_train, Y_train))
print(model.score(X_test, Y_test))

```

```python
import pickle

with open('random_forest_model.pkl', 'wb') as f:
    pickle.dump(model, f)
```
